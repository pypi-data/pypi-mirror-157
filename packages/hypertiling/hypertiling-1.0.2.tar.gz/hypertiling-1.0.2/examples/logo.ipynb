# Logo

```python
from hypertiling import HyperbolicTiling
from hypertiling.plot import plot_tiling, poly2patch

import matplotlib.pyplot as plt
import matplotlib.cm as cmap
import numpy as np
```

```python
p = 7
q = 3
nlayers = 3

T = HyperbolicTiling(p,q,nlayers) 
T.generate()

print("Number of polygons:", len(T))
```

```python
cT = np.zeros(len(T))
for i,t in enumerate(T):
    val = np.real(t.centerP)-np.imag(t.centerP)
    cT[i] = np.sign(val)*(np.abs(val))**1.4
```

```python
plot_tiling(T, cT, cmap=cmap.RdYlGn, edgecolor="w", lw=5, clim=[-1,1]);
```

```python
fig, ax = plt.subplots(figsize=(7,7), dpi=120)
pgons = poly2patch(T, cT, cmap=cmap.RdYlGn, edgecolor="w", lw=6, zorder=4)
ax.add_collection(pgons)

circ = plt.Circle((0, 0), radius=0.98, edgecolor='0.65', facecolor='0.92', zorder=1, lw=6)
ax.add_patch(circ)

plt.xlim([-1,1])
plt.ylim([-1,1])
plt.axis("off") 
plt.gca().set_aspect('equal');
```
